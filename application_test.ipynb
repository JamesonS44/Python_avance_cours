```python
# 1. Import
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# 2. CrÃ©er des donnÃ©es fictives
data = {
    "jour": pd.date_range(start="2026-01-01", periods=30),
    "Produit A": [10, 12, 15, 11, 14, 13, 10, 16, 18, 12, 14, 15, 13, 17, 16, 14, 15, 18, 20, 17, 15, 14, 13, 15, 17, 16, 18, 19, 14, 15],
    "Produit B": [5, 7, 6, 8, 9, 6, 5, 7, 8, 6, 7, 6, 8, 7, 9, 8, 7, 6, 5, 7, 6, 8, 7, 6, 5, 6, 7, 8, 7, 6],
    "Produit C": [20, 22, 18, 19, 21, 20, 23, 22, 21, 19, 20, 21, 22, 23, 24, 22, 20, 21, 23, 24, 22, 21, 20, 19, 21, 22, 23, 24, 25, 23]
}
df = pd.DataFrame(data)

# 3. CrÃ©er l'application
app = dash.Dash(__name__)

# 4. Layout : interface utilisateur
app.layout = html.Div([
    html.H1("ðŸ“Š Analyse des ventes par produit"),
    
    html.Label("SÃ©lectionnez un produit :"),
    dcc.Dropdown(
        id="dropdown-produit",
        options=[
            {"label": "Produit A", "value": "Produit A"},
            {"label": "Produit B", "value": "Produit B"},
            {"label": "Produit C", "value": "Produit C"}
        ],
        value="Produit A"
    ),
    
    html.Br(),
    html.Div(id="total-ventes", style={"fontSize": 20, "fontWeight": "bold"}),
    dcc.Graph(id="graph-ventes")
])

# 5. Callback : interactions
@app.callback(
    Output("graph-ventes", "figure"),
    Output("total-ventes", "children"),
    Input("dropdown-produit", "value")
)
def update_graph(produit):
    fig = px.bar(df, x="jour", y=produit, labels={"jour": "Jour", produit: "Ventes"})
    total = df[produit].sum()
    return fig, f"Total des ventes pour {produit} : {total}"

# 6. Lancer le serveur (nouvelle syntaxe Dash)
if __name__ == "__main__":
    app.run(debug=True)
```
